```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math as m



import warnings
warnings.filterwarnings("ignore")
```

```python
firearms = pd.read_csv("../Capstone/firearmMortatlity.csv")
```

```python
firearms
```

```python
pivot_firearms = pd.pivot_table(firearms,index = 'STATE', values = 'DEATHS', aggfunc ='mean')
```

```python
pivot_firearms
```

```python
pivot_firearms.plot( kind = 'bar')
plt.ylabel('DEATH')
plt.title('Average Deaths Per State')
plt.legend()
plt.show()
```

```python
TotalDeathState = firearms.groupby('STATE')['DEATHS'].sum().round(2)
print(TotalDeathState)
```

```python
Cali = firearms[firearms.STATE =='CA']
```

```python
FL = firearms[firearms.STATE =='FL']
```

```python
TX = firearms[firearms.STATE =='TX']
```

```python
plt.plot(Cali.YEAR, Cali.DEATHS)
plt.plot(FL.YEAR, FL.DEATHS)
plt.plot(TX.YEAR, TX.DEATHS)

plt.legend(['California', 'Florida', 'Texas'])
plt.xlabel('YEAR')
plt.ylabel('DEATHS')
plt.show()
```
